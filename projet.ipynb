### déroulement d'une journée:
-> on répartit l'eau du secteur \
-> on produit dans chaque unité de production, en vidant la source selon ce qui est consommé \
-> on répartit la production \
-> les gens consomment, et on calcule les points de vie\
-> on remplit la source selon son débit D et sa capacité 

# Résultats

Dans toutes les simulations suivantes, nous utilisons D = r = C. Ceci se justifie car après une rapide étude de cas, il apparaît que le paramètre C n'influe pas sur les résulats. D'autre part, les paramètres D et r influent de la même manière chacun. Afin de ne pas trop complexifier le projet qui vise à regarder différentes méthodes de répartition de l'eau entre les agents pour maximimiser le bien-être, nous faisons le choix de considérer D = r = C.

## Journée de type 1

**Rappel** : la journée de type 1 est telle que l'on donne autant d'eau à chaque agent ( fermes et gens ), donc il n'y a pas de priorité sur cette ressource. De plus, on donne autant de nourriture à chaque individu. \
Avec 1 ferme et 1 personne, on trouve un point de bascule entre un débit de 7 et un débit de 8. C'est-à-dire qu'en dessous de 8, l'agent va mourir de faim et au dessus de 8, on aura une croissance du bien-être au cours du temps, tout en évitant la mort de l'agent. \
**Remarque** : Ici l'individu meurt de faim et non de soif

```python
simulation_double(7,8, 7,8,7,8, 1,1,1,1,1,1, 100, "journée 1 D=C=r=7, 1 ferme, 1 agent", "journée 1 D=C=r=8, 1 ferme, 1 agent")
```

On essaie de généraliser le résultat trouvé précédemment et l'on observe qu'avec autant de fermes que d'agents ( N ), on trouve un point de bascule entre un débit de 7xN et un débit de 8xN. On a donc un cas identique avec 1 ferme et 1 agent mais à plus grande échelle

```python
simulation_double(799,800, 799,800,799,800, 100,100,100,100,1,1, 100, "journée 1 D=C=r=799, 100 fermes, 100 agents", "journée 1 D=C=r=800, 100 fermes, 100 agents")
```

On regarde maintenant ce qu'il se passe si le nombre de fermes est différent du nombre d'agents et plus particulièrement, on sait que le problème vient de la nourriture, donc on cherche le point de bascule.
\
Remarque : Avec ce modèle, on observe dans tous les graphes que les individus meurent d'un coup. 

```python
simulation_double(800,800, 800,800, 800,800, 99,100, 100,100, 1,1, 100, "journée 1 D=C=r=800, 99 fermes, 100 agents", "journée 1 D=C=r=800, 100 fermes, 100 agents")
```

Avec 10 fois moins de ressources, on remarque que l'on a un ratio du nombre de fermes/nombre de gens qui permet de ne pas être en situation critique \
ratio 60/15 = 4

```python
simulation_double(80,80, 80,80, 80,80, 59,60, 15,15, 1,1, 100, "journée 1 D=C=r=80, 59 fermes, 15 agents", "journée 1 D=C=r=80, 60 fermes, 15 agents")
```

En changeant les valeurs, tout en conservant un ratio de 4, on obtient toujours la même courbe avec le même point de bascule. \
Explication : La production par unité d'eau des fermes est 0.25 et les gens ont besoin d'une unité de nourriture quotidienne, donc pour que la nourriture soit produite en quantité suffisante, il faut que l'on ait 4 fois plus de fermes que d'agents.  

```python
simulation_double(80,80, 80,80, 80,80, 40,39, 10,10, 1,1, 100, "journée 1 D=C=r=80, 40 fermes, 10 agents", "journée 1 D=C=r=80, 39 fermes, 10 agents")
```

On regarde maintenant si l'on se met en situation critique où il y a beaucoup trop d'individus par rapport à la quantité de ressource disponible, tout en gardant un ratio de 4. Le but est de voir si les morts sont désormais de soif. \
Observation : Pour 40-10 (fermes-gens), on retrouve bien la courbe précédente où personne ne meurt et le bien-être croît. Pour 400-100 (fermes-gens), on a le même ratio mais beaucoup trop d'individus par rapport à la capacité en eau, donc le problème concerne la ressource en eau et les gens meurent de soif. 

```python
simulation_double(80,80, 80,80, 80,80, 400,40, 100,10, 1,1, 20, "journée 1 D=C=r=80, 400 fermes, 100 agents", "journée 1 D=C=r=80, 40 fermes, 10 agents")
```

On trouve un point de bascule entre 16 agents ( donc 64 fermes ) et 17 agents( donc 68 fermes ). C'est-à-dire qu'ici, la proportion de fermes/agents est suffisante pour que les gens ne meurent pas de faim. Mais dans le second cas, les ressources en eau ne sont pas suffisantes et l'on observe des morts. Néanmoins, on n'observe pas une mort complète de la population (cas traité juste après) puisque l'on voit qu'il reste 8 personnes vivantes. 

```python
simulation_double(80,80, 80,80, 80,80, 68,64, 17,16, 1,1, 100, "journée 1 D=C=r=80, 68 fermes, 17 agents", "journée 1 D=C=r=80, 64 fermes, 16 agents")
```

On s'intéresse au cas particulier cité précédemment : ici, on a un ratio tel que les fermes sont suffisamment nombreuses pour produire, donc les morts sont de soif. Entre 20 individus et 19 individus, on observe un comportement asymptotique complètement différent. Avec 80 fermes et 20 agents, tout le monde meurt, tandis qu'avec 76 fermes et 19 agents, on a d'abord une tendance à la mort de la majeure partie de la population, mais on a tout de même 4 survivants ! 

```python
simulation_double(80,80, 80,80, 80,80, 80,76, 20,19, 1,1, 100, "journée 1 D=C=r=80, 80 fermes, 20 agents", "journée 1 D=C=r=80, 76 fermes, 19 agents")
```

On plot en 3d pour t=100 le bien-être en fonction du nombre de gens et du nombre de fermes pour voir quel est l'optimum. Les paramètres fixés sont donc le temps ( on regarde à l'issue du 100ème jour ) et les capacité, débit, remplissage de la source d'eau. Les variables sont le nombre de gens et le nombre de fermes. 

```python
afficher(80, 80, 80, 50, 50, 1, 100, "Bien-être en fonction du nombre de fermes et de gens, t=100, D=r=C=80, journée 1")
```

En donnant plus de ressources en eau, on obtient un plot décalé vers la droite ( càd avec plus de fermes et plus de gens ), ce qui paraît cohérent avec le fait que l'on répartit équitablement. De surcroît, notre calcul de bien-être est calibré de manière à être croissant avec le nombre de gens vivants dans la population. Ce résultat est donc cohérent et la simulation sert de témoin. \

Dans les cas étudiés ici, la méthode la plus efficace de répartition de l'eau  (ie. celle débouchant au plus grand bien-être) était celle de la journée 2 donnant la priorité aux gens devant les fermes. Pour autant, avec les valeurs choisies, sans un ratio de 4 fois plus de fermes que de gens, ceux-ci auront du mal à tous survivre car ils n'auront pas assez  à manger.

Pistes d'enrichissement du modèle: les usines pourraient augmenter la productivité des fermes et de rajouter des points de vie si elles avaient elles aussi assez d'eau. Le sens derrière est de dire que si des usines peuvent développer de nouvelles technologies alors cela est profitable à la société en terme d'amélioration des facteurs de production mais également en terme de bien-être des gens ( apparitions de technologies de divertissement etc... ). Un problème qui aurait pu se poser en voulant maximiser le bien-être aurait été le capitalisme c'est-à-dire, il aurait peut-être été plus rentable de donner de l'eau aux usines plutôt qu'aux gens d'un point de vue de calcul de bien-être.
